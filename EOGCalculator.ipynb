```python
import sys
from tkinter import *
import tkinter as tk
import tkinter.font
from PIL import Image, ImageTk
import serial
import random

import matplotlib.pyplot as plt
import matplotlib as mpl

from time import sleep

try:
    arduino = serial.Serial("COM8",timeout=1);
except:
    print("Unable to connect to port")
```

```python
def clear():
    txtDisplay.delete(0,END);
    return;

#Parent Window.
root = Tk();
root.title('EOG Calculator');
root.geometry('900x830');

#Main entry.
number = StringVar();
txtDisplay = Entry(root, textvariable = number, relief=RIDGE, bd = 10, width=92, insertwidth = 1, font = 40);
txtDisplay.place(x=25, y=25);
txtDisplay.focus();

className = StringVar();
className.set("Biomedical System &")
classNameLabel = Label(root, textvariable = className, relief=RIDGE, width=20, bd=0);
classNameLabel.place(x=725, y=735);7

className1 = StringVar();
className1.set("Project Development")
classNameLabel1 = Label(root, textvariable = className1, relief=RIDGE, width=20, bd=0);
classNameLabel1.place(x=725, y=750);

user1 = StringVar();
user1.set("Ahmet Kaymak")
user1Label = Label(root, textvariable = user1, relief=RIDGE, width=20, bd=0);
user1Label.place(x=725, y=770);

user2 = StringVar();
user2.set("Yağmur Ercan")
user2Label = Label(root, textvariable = user2, relief=RIDGE, width=20, bd=0);
user2Label.place(x=725, y=785);

fsmvulogo = ImageTk.PhotoImage(Image.open("C:\\Users\\SAMSUNG\\Desktop\\EOGCalculator\\fsmvu_logo.png"))
imagePanel = Label(root, image = fsmvulogo);
imagePanel.place(x=760, y=635);

plusActive = minusActive = multiplyActive =  divideActive = False;
        
def update(value):
    currentValue = number.get()
    number.set(currentValue + value)
    
def add():
    global plusActive, minusActive, multiplyActive, divideActive;
    plusActive = True; minusActive = multiplyActive = divideActive = False;
    currentValue = number.get()
    number.set(currentValue + '+')

def minus():
    global plusActive, minusActive, multiplyActive, divideActive;
    minusActive = True; plusActive = multiplyActive = divideActive = False;
    currentValue = number.get()
    number.set(currentValue + '-')
    
def multiply():
    global plusActive, minusActive, multiplyActive, divideActive;
    multiplyActive = True; minusActive = plusActive = divideActive = False;
    currentValue = number.get()
    number.set(currentValue + 'x')
    
def divide():
    global plusActive, minusActive, multiplyActive, divideActive;
    divideActive = True; minusActive = multiplyActive = plusActive = False;
    currentValue = number.get()
    number.set(currentValue + '/')
    
def equal():
    currentValue = number.get()
    if(plusActive==True):
        [x,y]=currentValue.split('+')
        x=float(x); y=float(y);
        number.set(x+y)
    if(minusActive==True):
        [x,y]=currentValue.split('-')
        x=float(x); y=float(y);
        number.set(x-y)
    if(multiplyActive==True):
        [x,y]=currentValue.split('x')
        x=float(x); y=float(y);
        number.set(x*y)
    if(divideActive==True):
        [x,y]=currentValue.split('/')
        x=float(x); y=float(y);
        number.set(x/y)

startButton = Button(root, text='Start', width=20, height=10, bg='white', fg='black',command = lambda: change()).place(x=550,y=645);
        
#Buttons:
zeroButton = Button(root, text='0', width=70, height=10, bg='white', fg='black',command = lambda: update('0'));
zeroButton.place(x=25,y=645);

oneButton = Button(root, text='1', width=20, height=10, bg='white', fg='black', command = lambda: update('1'));
oneButton.place(x=25, y=460);

twoButton = Button(root, text='2', width=20, height=10, bg='white', fg='black', command = lambda: update('2'));
twoButton.place(x=200, y=460);

threeButton = Button(root, text='3', width=20, height=10, bg='white', fg='black', command = lambda: update('3'));
threeButton.place(x=375, y=460);

fourButton = Button(root, text='4', width=20, height=10, bg='white', fg='black', command = lambda: update('4'));
fourButton.place(x=25, y=275);

fiveButton = Button(root, text='5', width=20, height=10, bg='white', fg='black', command = lambda: update('5'));
fiveButton.place(x=200, y=275);

sixButton = Button(root, text='6', width=20, height=10, bg='white', fg='black', command = lambda: update('6'));
sixButton.place(x=375, y=275);

sevenButton = Button(root, text='7', width=20, height=10, bg='white', fg='black', command = lambda: update('7'));
sevenButton.place(x=25, y=90);

eightButton = Button(root, text='8', width=20, height=10, bg='white', fg='black', command = lambda: update('8'));
eightButton.place(x=200, y=90);

ninthButton = Button(root, text='9', width=20, height=10, bg='white', fg='black', command = lambda: update('9'));
ninthButton.place(x=375, y=90);

plusButton = Button(root, text='+', width=20, height=10, bg='orange', fg='black', command = lambda: add());
plusButton.place(x=550, y=90);

minusButton = Button(root, text='-', width=20, height=10, bg='orange', fg='black', command = lambda: minus());
minusButton.place(x=725, y=90);

multiplyButton = Button(root, text='x',width=20, height=10, bg='orange', fg='black', command = lambda: multiply());
multiplyButton.place(x=550, y=275);

divideButton = Button(root, text='÷', width=20, height=10, bg='orange', fg='black', command = lambda: divide());
divideButton.place(x=725, y=275);

equalButton = Button(root, text='=', width=20, height=10, bg='Lightgreen', fg='black', command = lambda: equal());
equalButton.place(x=550, y=460);

clearButton = Button(root, text='Clear (CE)', width=20, height=10, bg='Lightgreen', fg='black', command = clear);
clearButton.place(x=725, y=460);

def change():
    global zeroButton,oneButton,twoButton,threeButton,fourButton,fiveButton,sixButton,sevenButton,eightButton,ninthButton;
    global plusButton,minusButton,multiplyButton,divideButton,equalButton,clearButton;
    sixButton.config(bg="yellow");
    currentValue="6";
    while(True):
        RLValue=int(str(arduino.readline()).split("'")[1].split("\\")[0]);
        if(RLValue!=""):
            # All right cases
            if(RLValue>300):
                if(currentValue=="1"):
                    currentValue="2";
                    oneButton.config(bg="white");
                    twoButton.config(bg="yellow");
                if(currentValue=="2"):
                    currentValue="3";
                    twoButton.config(bg="white");
                    threeButton.config(bg="yellow");
                if(currentValue=="3"):
                    currentValue="=";
                    threeButton.config(bg="white");
                    equalButton.config(bg="yellow");
                if(currentValue=="="):
                    currentValue="ce";
                    equalButton.config(bg="white");
                    clearButton.config(bg="yellow");
                if(currentValue=="4"):
                    currentValue="5";
                    fourButton.config(bg="white");
                    fiveButton.config(bg="yellow");
                if(currentValue=="5"):
                    currentValue="6";
                    fiveButton.config(bg="white");
                    sixButton.config(bg="yellow");
                if(currentValue=="6"):
                    currentValue="x";
                    sixButton.config(bg="white");
                    multiplyButton.config(bg="yellow");
                if(currentValue=="x"):
                    currentValue="÷";
                    multiplyButton.config(bg="white");
                    divideButton.config(bg="yellow");
                if(currentValue=="7"):
                    currentValue="8";
                    sevenButton.config(bg="white");
                    eightButton.config(bg="yellow");
                if(currentValue=="8"):
                    currentValue="9";
                    eightButton.config(bg="white");
                    ninthButton.config(bg="yellow");
                if(currentValue=="9"):
                    currentValue="+";
                    ninthButton.config(bg="white");
                    plusButton.config(bg="yellow");
                if(currentValue=="+"):
                    currentValue="-";
                    plusButton.config(bg="white");
                    minusButton.config(bg="yellow");
                root.update_idletasks();
                root.update();
            # All left cases
            if(RLValue<0):
                if(currentValue=="ce"):
                    currentValue="=";
                    clearButton.config(bg="white");
                    equalButton.config(bg="yellow");
                if(currentValue=="="):
                    currentValue="3";
                    equalButton.config(bg="white");
                    threeButton.config(bg="yellow");
                if(currentValue=="3"):
                    currentValue="2";
                    threeButton.config(bg="white");
                    twoButton.config(bg="yellow"); 
                if(currentValue=="2"):
                    currentValue="1";
                    twoButton.config(bg="white");
                    oneButton.config(bg="yellow");
                if(currentValue=="÷"):
                    currentValue="x";
                    divideButton.config(bg="white");
                    multiplyButton.config(bg="yellow");
                if(currentValue=="x"):
                    currentValue="6";
                    multiplyButton.config(bg="white");
                    sixButton.config(bg="yellow");
                if(currentValue=="6"):
                    currentValue="5";
                    sixButton.config(bg="white");
                    fiveButton.config(bg="yellow");
                if(currentValue=="5"):
                    currentValue="4";
                    fiveButton.config(bg="white");
                    fourButton.config(bg="yellow");
                if(currentValue=="-"):
                    currentValue="+";
                    minusButton.config(bg="white");
                    plusButton.config(bg="yellow");
                if(currentValue=="+"):
                    currentValue="9";
                    plusButton.config(bg="white");
                    ninthButton.config(bg="yellow");
                if(currentValue=="9"):
                    currentValue="8";
                    ninthButton.config(bg="white");
                    eightButton.config(bg="yellow");
                if(currentValue=="8"):
                    currentValue="7";
                    eightButton.config(bg="white");
                    sevenButton.config(bg="yellow");
                root.update_idletasks();
                root.update();   
            
            
root.maxsize(900,830);
root.minsize(900,830);
root.mainloop();
```
